```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Carregar o arquivo CSV no pandas
df = pd.read_csv('ecommerce_preparados.csv')

# Verificar as primeiras linhas do dataset
df.head()
```

```python
# Informações gerais sobre o dataframe (tipos de dados, valores nulos, etc.)
df.info()

# Estatísticas descritivas das variáveis numéricas
df.describe()
```

```python
# Exibir o nome de todas as colunas do DataFrame
df.columns
```

```python
# Histograma 1
plt.figure(figsize=(10,6))
df['N_Avaliações'].hist(bins=30, edgecolor='black')
plt.title('Histograma de [N_Avaliações]')
plt.xlabel('[N_Avaliações]')
plt.ylabel('Frequência')
plt.show()
```

```python
# Gráfico de Dispersão 1
plt.figure(figsize=(10,6))
plt.scatter(df['N_Avaliações'], df['Nota'])
plt.title('Gráfico de Dispersão entre [N_Avaliações] e [Nota]')
plt.xlabel('[N_Avaliações]')
plt.ylabel('[Nota]')
plt.show()
```

```python
# Selecionando apenas as colunas numéricas
numerical_df = df.select_dtypes(include=np.number)

# Calculando a matriz de correlação apenas para as colunas numéricas
corr = numerical_df.corr()

# Mapa de calor
plt.figure(figsize=(10,6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Mapa de Calor das Correlações')
plt.show()
```

```python
from logging import StreamHandler
# Gráfico de densidade
plt.figure(figsize=(10,6))
sns.kdeplot(df['Desconto_MinMax'], shade=True)
plt.title('Gráfico de Densidade de [Desconto_MinMax]')
plt.xlabel('[Desconto_MinMax]')
plt.ylabel('Densidade')
plt.show()
```

```python
# Gráfico de regressão
plt.figure(figsize=(10,6))
sns.regplot(x='Preço_MinMax', y='Desconto_MinMax', data=df, scatter_kws={'color':'blue'}, line_kws={'color':'red'})
plt.title('Gráfico de Regressão entre [Preço_MinMax] e [Desconto_MinMax]')
plt.xlabel('[Preço_MinMax]')
plt.ylabel('[Desconto_MinMax]')
plt.show()
```
